# Statistik Tentang Data

```python
sum = data_ipm.angka_harapan_hidup.sum()
mean = data_ipm.angka_harapan_hidup.mean()
min = data_ipm.angka_harapan_hidup.min()
max = data_ipm.angka_harapan_hidup.max()

print('=== Total, Rata-Rata, Min, Max Angka Harapan Hidup  ===')
print('Total : ', sum)
print('Rata2 : ', mean)
print('Min : ', min)
print('Max : ', max, '\n')

sum = data_ipm.angka_melek_huruf.sum()
mean = data_ipm.angka_melek_huruf.mean()
min = data_ipm.angka_melek_huruf.min()
max = data_ipm.angka_melek_huruf.max()

print('=== Total, Rata-Rata, Min, Max Angka Melek Huruf  ===')
print('Total : ', sum)
print('Rata2 : ', mean)
print('Min : ', min)
print('Max : ', max, '\n')

sum = data_ipm.lama_sekolah.sum()
mean = data_ipm.lama_sekolah.mean()
min = data_ipm.lama_sekolah.min()
max = data_ipm.lama_sekolah.max()

print('=== Total, Rata-Rata, Min, Max Lama Sekolah  ===')
print('Total : ', sum)
print('Rata2 : ', mean)
print('Min : ', min)
print('Max : ', max, '\n')

sum = data_ipm.pengeluaran_perkapita.sum()
mean = data_ipm.pengeluaran_perkapita.mean()
min = data_ipm.pengeluaran_perkapita.min()
max = data_ipm.pengeluaran_perkapita.max()

print('=== Total, Rata-Rata, Min, Max Pengeluaran Perkapita  ===')
print('Total : ', sum)
print('Rata2 : ', mean)
print('Min : ', min)
print('Max : ', max, '\n')

sum = data_ipm.ipm.sum()
mean = data_ipm.ipm.mean()
min = data_ipm.ipm.min()
max = data_ipm.ipm.max()

print('=== Total, Rata-Rata, Min, Max IPM  ===')
print('Total : ', sum)
print('Rata2 : ', mean)
print('Min : ', min)
print('Max : ', max, '\n')
```

# Visualisasi Sederhana

# 1. Rata-rata IPM Tiap Provinsi dari Tahun 2004-2012

```python
ax = data_ipm.groupby('nama_provinsi').mean()['ipm']
ax = ax.plot(kind='bar', title ="Rata-Rata IPM Tiap Provinsi", figsize=(15,10), legend=True, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("IPM", fontsize=12)
```

# 2. Rata-rata IPM Tiap Tahun (2004-2012)

```python
ax = data_ipm.groupby('tahun').mean()['ipm']
ax = ax.plot(kind='line', title ="Rata-Rata IPM Tiap Tahun", figsize=(7,10), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("IPM", fontsize=12)
```
